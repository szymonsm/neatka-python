```python
import pandas as pd

df_kan_neat_summary = pd.read_json("results/all_test_results_summary.json")
# select columns: function_name, seed, train_rmse, test_rmse, num_generations
df_kan_neat_summary = df_kan_neat_summary[["function_name", "seed", "train_rmse", "test_rmse"]]
df_kan_neat_summary
```

```python
# rename function_name to function
df_kan_neat_summary.rename(columns={"function_name": "function"}, inplace=True)
```

```python
df_all_results = pd.read_csv("special_functions_results/experiment_results.csv")
df_all_results
```

```python
# rename train_loss and test_loss to train_rmse and test_rmse
df_all_results.rename(columns={"train_loss": "train_rmse", "test_loss": "test_rmse"}, inplace=True)
```

```python
# Merge the relevant part of df_all_results with df_kan_neat_summary
mask = df_all_results["method"] == "kan-neat"
df_to_update = df_all_results[mask].merge(
    df_kan_neat_summary[["seed", "function", "train_rmse", "test_rmse"]],
    on=["seed", "function"],
    how="left",
    suffixes=("", "_new")
)

# Update values in df_all_results where applicable
df_all_results.loc[mask, "train_rmse"] = df_to_update["train_rmse_new"].values
df_all_results.loc[mask, "test_rmse"] = df_to_update["test_rmse_new"].values
```

```python
df_all_results
```

```python
# select function, method, train_rmse, test_rmse, seed
df_all_results_selected = df_all_results[["function", "method", "train_rmse", "test_rmse", "seed"]]
df_all_results_selected
```

```python
# Select best combinations grouped by function and method based on test_rmse
df_best_results = df_all_results_selected.loc[df_all_results_selected.groupby(["function", "method"])["test_rmse"].idxmin()]
df_best_results
```

```python
# save best results to a new CSV file
df_best_results.to_csv("best_results_per_function_method.csv", index=False)
```
